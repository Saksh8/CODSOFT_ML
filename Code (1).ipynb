## Imports

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

## Reading the dataset 

```python
data=pd.read_csv("creditcard.csv")
data
```

## Data Information

```python
data.info()
```

## Data Preprocessing

#### Handling Missing Values

```python
data.isnull().sum()  #data doesnot contain null values
```

#### Data Distribution

```python
data['Class'].value_counts()  #it is a unbalanced data since more than 70% of values are of a particular type
```

```python
#seperating the true and false transaction
#0: True Transactions
#1: False Transaction

true_tr=data[data['Class']==0]
false_tr=data[data['Class']==1]
```

```python
print(true_tr.shape)
print(false_tr.shape)
```

```python
#statistical measure of the data
true_tr['Amount'].describe()
```

```python
false_tr['Amount'].describe()
```

```python
#comparing the mean values of the two type of transaction
data.groupby("Class").mean()
```

## Data Sampling

We will begin doing the undersampling of the given dataset which means building a sample dataset containing similar distribution of true_transaction and false transaction
Number of false_transaction=492


```python
true_tr=true_tr.sample(492)
```

Concatenation of two dataset

```python
new_data=pd.concat([true_tr,false_tr],axis=0)
```

```python
new_data['Class'].value_counts()
```

#### Feature Selection

```python
X=new_data.drop(columns='Class')
Y=new_data['Class']
```

```python
X
```

```python
Y
```

#### Train Test Split 

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
X_train
```

## Model Training-Logistic Regression

```python
lr=make_pipeline(StandardScaler(),LogisticRegression())
lr.fit(X_train,Y_train)
```

## Model Evaluation 

```python
#on training Data
Y_pred=lr.predict(X_train)
Model_Accuracy=accuracy_score(Y_train,Y_pred)
Model_precision=precision_score(Y_train,Y_pred)
Model_recall=recall_score(Y_train,Y_pred)
Model_F1_score=f1_score(Y_train,Y_pred)
```

```python
print("Accuracy:",Model_Accuracy)
```

```python
print("Precision:",Model_precision)
```

```python
print("Recall:",Model_recall)
```

```python
print("F1_Score",Model_F1_score)
```

```python
#on testing Data
Y_pred=lr.predict(X_test)
Model_Accuracy=accuracy_score(Y_test,Y_pred)
Model_precision=precision_score(Y_test,Y_pred)
Model_recall=recall_score(Y_test,Y_pred)
Model_F1_score=f1_score(Y_test,Y_pred)
```

```python
print("Accuracy:",Model_Accuracy)
```

```python
print("Precision:",Model_precision)
```

```python
print("Recall:",Model_recall)
```

```python
print("F1_Score",Model_F1_score)
```
